## 2) Select Model

```python
# make a function 'line_model' that models y = m*x + b
# the function takes in:
# x - an x value, and
# theta - where theta[0] will be 'm' and theta[1] is 'b'
# the functions returns:
# y = m*x + b
def line_model(x,theta):
    m = theta[0]
    b = theta[1]
    y = m*x + b
    return y
```

```python
#pick some values for x, m and b
# use them to make theta
# check to make sure line_model() works
x = 3
array = [3, 6]
temp_line_model = line_model(x, array)
print("x was " + str(x) + " and y was " + str(temp_line_model))
```

## 3) Select Cost Function

```python
# Note this cost function is based on absolute value
def cost_function(theta,args):
    xs = args[0]
    ys = args[1]
    m2 = theta[0] # I think this wasn't scoped properly, so I declared it.
    b = theta[1] # I think this wasn't scoped properly, so I declared it.
    print(m2,b,xs,ys)
    cost = 0
    for n in range(len(xs)):
        cost += np.abs(line_model(xs[n],theta)-ys[n]) 
    return cost
```

```python
from scipy.optimize import minimize
```

```python
# initial (bad) guess at (x,y) values
initial_guess = np.array([0,0])
args = [xs,ys]
cost_val = cost_function(initial_guess,args)
print(cost_val)

#TODO WHY IS THIS SO MESSY? I WANT A SINGLE COST NUMBER OUTPUT NOT AN ARRAY.
```

```python
opt_result = minimize(cost_function,initial_guess,args = [xs,ys])

#AGAIN WITH THE MESSY. I WANT A SINGLE COST NUMBER FOR EACH GUESS.
```

```python
# get results of optimization
print(opt_result.x)
m_best = opt_result.x[0]    #where are these attribubtes on opt_result coming from? 
b_best = opt_result.x[1]        #We never declared an optr_result.x variable.
theta_best = [m_best,b_best]
```

## Try it all over with a 5th degree polynomial model!

```python
# I know I don't have to import these again. It's so I can quickly see the variables.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# GET THE DATA
df = pd.read_csv("fake-data.txt")
xVals = df["x1"].to_numpy()
yVals = df["y"].to_numpy()

# CHOOSE THE MODEL
# y = ax^5 + bx^4 + cx^3 + dx^2 + ex + f
def line_model(x, theta):
    
    a = theta[0]
    b = theta[1]
    c = theta[2]
    d = theta[3]
    e = theta[4]
    f = theta[5]
    
    y = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f
    
    return y

# DEFINE THE COST FUNCTION AND FIND BEST WEIGHTS
def cost_finder(testTheta, valueArray):
   
    xArray = valueArray[0]
    yArray = valueArray[1]
    theta = testTheta
    cost = 0
    
    for i in range(len(xArray)):
        
        cost += np.abs(line_model(xArray[i], theta)-yArray[i])
     
    return cost

# FIND MINIMIIZED FUNCTION THETA VALUES
first_guess = [0, 0, 0, 0, 0, 0]

#Find the best theta values
best_theta_vals = minimize(cost_finder, first_guess, args = [xVals, yVals])

# magically extract the values into best_theta
best_theta = []
for i in range(0, len(best_theta_vals.x)):
    
    best_theta.append(best_theta_vals.x[i])

# PLOT THE ORIGIONAL FUNCTION AND A PLOT OF THE NEW REGRESSION MODEL
plt.scatter(xVals, yVals)

yGuess = []
for n in range(0, len(xVals)):
    
    newY = line_model(xVals[n], best_theta)
    yGuess.append(newY)
    
#plot the guess 5'th degree polynomial.
plt.plot(xVals, yGuess, "b")    
```
