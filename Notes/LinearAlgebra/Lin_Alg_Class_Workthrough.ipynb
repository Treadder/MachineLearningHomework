```python
# calculate the length of vector 'b' (from above) using the best function         <--Done!

def calculate_vector_length(vector):
    
    print("Starting to calculate length!")
    print("input vector is" + str(vector))
    
    sum = 0
    
    for i in range(0, len(vector)):
        sum+=vector[i]**2
        print("i is: " + str(i) + " our value add/square is " + str(vector[i]) + " and sum is: " + str(sum))
        
        
    sum = np.sqrt(sum)
    print("time to take square root! final answer is:" + str(sum))
    
    return sum

calculate_vector_length(b)

# then multiply each element in 'b' by a factor of 3 and recalculate the length          <--Done!

b = b*3
print("")
print("multiplied vector by 3")
print("")
calculate_vector_length(b)
```

```python
# normalize 'b' and call it 'b_hat'        <--Done!

def normalize_vector(vector):
    
    magnitude = calculate_vector_length(vector)
    print("Magnitude is: " + str(magnitude))
    
    print("Dividing inpur vector by magnitude:")
    unit_vector = vector/magnitude
    print("unit vector is: " + str(unit_vector))
    
    return(unit_vector)

b_hat = normalize_vector(b)
print("b_hat is:" + str(b_hat))

# calculate the length of 'b_hat' Note: sometimes it will be like 0.9999999999999999 but that's pretty normalized.
b_hat_length = calculate_vector_length(b_hat)
print("b_hat length is: " + str(b_hat_length))
```

```python
# calculate the angle between 'a' and 'b'          <--Done!

#just a quick placeholder.
a = a
b = b

def dot_product_of_linear_vectors(a, b):
    count = 0
    for i in range(0, len(a)):
        count+=(a[i]*b[i])
    return count

def calculate_angle_between_vectors(a, b):
    magnitude_a = calculate_vector_length(a)
    magnitude_b = calculate_vector_length(b)
    a_dot_b = dot_product_of_linear_vectors(a,b)
    
    frac = (a_dot_b/(magnitude_a*magnitude_b))
    angle_between = np.arccos(frac)
    print("Angle between vectors: " + str(frac))
    return frac

angle = calculate_angle_between_vectors(a, b)
print(angle)
    
# find the projection of 'a' onto 'b' (NOTE: this is a vector)         <--Done!

#Finds the projection of shadowMaker onto base
def find_projection(base, shadowMaker):
    
    base_unit_vector = normalize_vector(base)
    numerator = dot_product_of_linear_vectors(shadowMaker, base_unit_vector)
    denominator = dot_product_of_linear_vectors(base_unit_vector, base_unit_vector)
    multiplier = base_unit_vector
    
    projection = (numerator/denominator)*multiplier
    
    print("projection: " + str(projection))
    return projection

projection = find_projection(b, a)
print(projection)
```

```python
# multiply every element in 'm' by 4           <--Done!
matrix2 = m.copy() #make a backup of m so I don't mess it up in this cell

def multiply_matrix_by_scalar(matrix, scalar):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[0])):
            matrix[i][j] = scalar*matrix[i][j]
    return matrix;

multiplied_matrix = multiply_matrix_by_scalar(matrix2, 4)

print("matrix:")
print(m)
print("multiplied matrix:")
print(multiplied_matrix)
```

```python
# find the transpose of 'm'                 <---Done!

def transpose_matrix(matrix):
    returnMatrix = np.zeros([len(matrix[0]), len(matrix)])
    for i in range(0, len(matrix)):
        returnMatrix[:,i] = matrix[i]
    return returnMatrix

m_transpose = transpose_matrix(m)

print("m is:")
print(m)
print("")

print("m_transpose is:")
print(m_transpose)
print("")
```
