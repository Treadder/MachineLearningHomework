```python
# create 2 3-D vectors, called 'a' and 'b', with random (real) values in them ranging between 0 and 10
a = np.random.randint(low=0, high=10, size=3)
print("a:")
print(a)

b = np.random.randint(low=0, high=10, size=3)
print("b:")
print(b)

# use 'zip' to combine them and extract the x,y, and z coords
c = np.array(list(zip(a, b)))#Because python likes to be annoying, you have to wrap a zipped array in a list.
print("zipped array:")
print(c)

# plot the 2 points in a 3-D scatter plot

fig = plt.figure()
fig = Axes3D(fig)
fig.scatter(c[0], c[1], zs=c[2])#TODO make some lines show up.
```

```python
# calculate the length of vector 'b' (from above) using the best function

def calculate_vector_length(vector):
    
    print("input vector is" + str(vector))
    
    sum = 0
    
    for i in range(0, len(vector)):
        print("i is: " + str(i) + " our value add/square is " + str(vector[i]) + " and sum is: " + str(sum))
        sum+=vector[i]**2
        
    sum = np.sqrt(sum)
    print("time to take square root! final answer is:")
    
    return sum

calculate_vector_length(b)





# then multiply each element in 'b' by a factor of 3 and recalculate the length
```
